# Import required libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Step 1: Load the dataset

```python

iris = sns.load_dataset('iris')
```


# Step 2: Inspect the dataset

```python
print("Shape of the dataset:", iris.shape)
print("\nColumn names:", iris.columns.tolist())
print("\nFirst five rows:")
print(iris.head())
```


# Step 3: Dataset info and summary statistics

```python
print("\nDataset Info:")
print(iris.info())

print("\nSummary Statistics:")
print(iris.describe())
```

# Step 4: Data Visualization

## 4.1 Scatter Plot â€” relationship between features

```python
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species', palette='viridis')
plt.title('Scatter Plot: Sepal Length vs Petal Length')
plt.show()
```

## 4.2 Histograms â€” feature distributions

```python

iris.hist(figsize=(10, 8), bins=20, edgecolor='black')
plt.suptitle('Histograms of Iris Features', fontsize=14)
plt.show()
```

## 4.3 Box Plots â€” identifying outliers

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris)
plt.title('Box Plots of Iris Features')
plt.show()
```


## 4.4 Pair Plot â€” relationships among all features

```python
sns.pairplot(iris, hue='species', palette='husl')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```
